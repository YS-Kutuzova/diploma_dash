```python
import pandas as pd
import numpy as np

# Загрузка датасета
df = pd.read_csv('all_sport_objects_with_infrastructure.csv')
df
```

```python
# Создаем функцию для проверки условий
def check_schedule(schedule_value):

    if pd.isna(schedule_value):
        return 0

    schedule_str = str(schedule_value)

    has_00_00 = "00:00" in schedule_str
    has_to_24_00 = '"to": "24:00"' in schedule_str or "'to': '24:00'" in schedule_str

    return 1 if (has_00_00 and has_to_24_00) else 0

df['schedule'] = df['schedule'].apply(check_schedule)
```

```python
def update_sport_type(row):
    keyword = row['search_keyword']
    if keyword == 'теннисный корт':
        return 'теннис'
    elif keyword == 'сквош корт':
        return 'сквош'
    elif keyword == 'падел':
        return 'паддл'
    else:
        # Оставляем текущее значение
        return row['sport_object_type']

df['sport_object_type'] = df.apply(update_sport_type, axis=1)
df
```

# Гипотезы:

### 1) Районы с более высокой плотностью населения имеют лучшее покрытие спортивными объектами.

### 2) Районы с более высокой средней зарплатой имеют лучшее покрытие спортивными объектами.

### 3) Районы с более высокой плотностью объектов инфраструктуры (кафе, ТЦ, супермаркеты, офисы, остановки и тд) имеют большее количество спортивных объектов.

### 4) Районы с высоким соотношением мужчин к женщинам (Соотношение_М_Ж) могут иметь больше количество спортивных объектов.



# Аналитика по гипотезам

```python
# Группируем данные по районам для анализа

district_stats = df_final.groupby('district').agg({
    'sport_object_id': 'nunique',
    'sport_object_type': 'nunique',
    'Плотность_населения': 'first',
    'Зарплата': 'first',
    'Соотношение_М_Ж': 'first',
    'Кластер': 'first',
    'Класс_плотности': 'first',
    'Класс_зарплаты': 'first'
}).reset_index()

district_stats = district_stats.rename(columns={
    'sport_object_id': 'sport_objects_count',
    'sport_object_type': 'sport_types_count'
})
district_stats
```

```python
# Рассчитываем плотность инфраструктуры по районам

# Количество объектов инфраструктуры по районам
infra_by_district = df_final.groupby('district')['infrastructure_id'].nunique().reset_index()
infra_by_district = infra_by_district.rename(columns={'infrastructure_id': 'infrastructure_count'})

# Объединяем с основными данными
district_stats = pd.merge(district_stats, infra_by_district, on='district', how='left')

# Рассчитываем плотность (инфраструктура на 1 спортивный объект)
district_stats['infrastructure_per_sport_object'] = (
    district_stats['infrastructure_count'] / district_stats['sport_objects_count']
).round(2)  # округлим
```

```python
# Рассчитываем показатели на душу населения
district_stats['sport_objects_per_100k'] = (
    district_stats['sport_objects_count'] / district_stats['Население'] * 100000
).round(2)

district_stats['infrastructure_per_100k'] = (
    district_stats['infrastructure_count'] / district_stats['Население'] * 100000
).round(2)

# Выводим статистику по подготовленным данным
print(f"Среднее количество спортивных объектов на район: {district_stats['sport_objects_count'].mean():.1f}")
print(f"Среднее количество видов спорта на район: {district_stats['sport_types_count'].mean():.1f}")
```
